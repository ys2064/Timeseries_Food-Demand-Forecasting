# Validationset: Actual vs Predicted weekly orders

```python
import pandas as pd
predictions_valset = '/content/drive/My Drive/TCN/Validationset Weekly Order Prediction.xlsx'
df = pd.read_excel(predictions_valset)
print(df)
```

```python
import matplotlib.pyplot as plt
predictions_valset = '/content/drive/My Drive/TCN/Validationset Weekly Order Prediction.xlsx'
df = pd.read_excel(predictions_valset)
plt.figure(figsize=(12, 6))
plt.plot(df["week"], df["total_num_orders_winsorized"], label="Actual", marker='o', linestyle='-')
plt.plot(df["week"], df["predicted_num_of_orders"], label="Predicted", marker='x', linestyle='--')
plt.xlabel("Week", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.title("Valset: Actual vs Predicted Number of Orders", fontsize=14)
plt.legend(fontsize=12)
plt.xticks(df["week"], rotation=45)
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
